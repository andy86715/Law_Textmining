```python
import re
regex = re.compile('民法第1052條第1項第?(\d+(?:、\d+)*)款')
regex2 = re.compile('民法第1052條第1項第款')
regex3 = re.compile('民法第1052條第2項')
regex4 = re.compile('民法第1052條第1項以外之重大事由')
regex5 = re.compile('[民同]法第1052條第(\d+、*\d*)款')
```

```python
law = []
for i,k in list(zip(df['content'],df['Id'])):
    match = regex.search(i)
    match2 = regex2.search(i)
    match3 = regex3.search(i)
    match4 = regex4.search(i)
    match5 = regex5.search(i)
    try:
        if '第' not in match.group(0):
            law.append(match.group(0).split("項")[0]+"項第"+match.group(0).split("項")[1])
        else:
            law.append(match.group(0))
    except:
        try:
            match2.group(0)
            law.append('民法第1052條第1項第10款')
        except:
            try:
                law.append(match3.group(0))
            except:
                try:
                    law.append(match4.group(0))
                except:
                    try:
                        if '同' in match5.group(0):
                            string = match5.group(0).replace("同","民")
                            law.append(string.split('條')[0]+'條'+'第1項'+string.split('條')[1])
                        else:
                            law.append(match5.group(0).split('條')[0]+'條'+'第1項'+match5.group(0).split('條')[1])
                    except:
                        law.append(k)
```
